# 2) สำรวจค่าสูญหาย
	- ค่าสูญหายถูกแทนด้วยสัญลักษณ์อะไร?
	- มีค่าสูญหายจำนวนเท่าไหร่ในแต่ละคอลัมน์
	- ในกรณีนี้ให้แทนค่าสูญหายด้วย mean/median/mode

```python
adult_data.isnull().sum()
```

```python
adult_data[['workclass']] = adult_data[['workclass']].fillna(adult_data['workclass'].mode()[0])
adult_data[['occupation']] = adult_data[['occupation']].fillna(adult_data['occupation'].mode()[0])
adult_data[['native-country']] = adult_data[['native-country']].fillna(adult_data['native-country'].mode()[0])
```

```python
adult_data.isnull().sum()
```

# **3) ตอบคำถาม**
	- ตัวอย่างมีถิ่นกำเนิดที่ประเทศใดบ้าง? มีคนไทยกี่คน? #กุบบาย>เค้า
	- สร้างตัวแปร capital-total = capital-gain - capital-loss
	- ตัวอย่างที่ถิ่นกำเนิดใดมีค่าเฉลี่ยของ capital-total มากที่สุด
	- อาชีพใด (occupation) มีจำนวนตัวอย่างที่มีรายได้ >50K มากที่สุด #สร้างตารางสรุป



```python
set(adult_data['native-country'])
```

```python
adult_data[adult_data['native-country'] == 'Thailand']['native-country'].count()
```

```python
adult_data['capital-total'] = adult_data['capital-gain'] - adult_data['capital-loss']
```

```python
#ตัวอย่างที่ถิ่นกำเนิดใดมีค่าเฉลี่ยของ capital-total มากที่สุด
adult_data.groupby(['native-country'])[['capital-total']].mean().sort_values(by='capital-total')
```

```python
#อาชีพใด (occupation) มีจำนวนตัวอย่างที่มีรายได้ >50K มากที่สุด #สร้างตารางสรุป
occupation = adult_data[adult_data['income'] == '>50K']['occupation'].value_counts()
occupation
```

```python
occupation = adult_data[adult_data['income'] == '>50K'].groupby('occupation').size()
occupation_max = occupation.idxmax()

print("อาชีพที่มีจำนวนตัวอย่างที่มีรายได้ >50K มากที่สุด คือ ", occupation_max)
```
