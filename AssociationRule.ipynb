```python
import pandas as pd
import numpy as np
# do Apriori in Python
from mlxtend.frequent_patterns import apriori, association_rules
import apyori
```

```python
# Creating Sample Transactions
transactions = [
                 ['Milk', 'Bread', 'Saffron'],
                 ['Milk', 'Saffron'],
                 ['Bread', 'Saffron','Wafer'],
                 ['Bread','Wafer'],
               ]
```

```python
# Generating association rules
Rules = list(apyori.apriori(transactions,
                            min_support=0.5,
                            min_confidence=0.5))
```

```python
# Extracting rules from the object
for i in range(len(Rules)):
    LHS = list(Rules[i][2][0][0])
    RHS = list(Rules[i][2][0][1])
    support = Rules[i][1]
    confidence = Rules[i][2][0][2]
    lift = Rules[i][2][0][3]
    print("LHS:",LHS,"--","RHS:",RHS)
    print("Support:",support)
    print("Confidence:",confidence)
    print("Lift:",lift)
    print(10*"----")
```

### Apply Apriority using mlxtend

#### Create an incident matrix

```python
# Let's transform the list, with one-hot encoding
from mlxtend.preprocessing import TransactionEncoder
a = TransactionEncoder()
a_data = a.fit(data).transform(data)
sl = pd.DataFrame(a_data,columns=a.columns_)
sl = sl.replace(False,0)
sl
```

```python
a_data
```

```python
#set a threshold value for the support value and calculate the support value.
frequent_itemsets = apriori(sl, min_support = 0.2, use_colnames = True, verbose = 1)
frequent_itemsets
```

```python
frequent_support = apriori(sl, min_support = 0.3, use_colnames = True, verbose = 1)
frequent_support
```

```python
# Let's view our interpretation values using the Associan rule function.
frequent_conf = association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.6)
frequent_conf
```

```python
frequent_conf.iloc[1]#ซื้อ SUGERและBREAD = 0.2
```
